```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('mpg(1)(2).csv')
data.head()
```

```python
plt.figure()

sns.kdeplot(data.loc[data.cyl==5,'cty'],shade=True,color='green',label='Cyl=4')
sns.kdeplot(data.loc[data.cyl==4,'cty'],shade=True,color='yellow',label='Cyl=4')

plt.title('Density Plot')
```

```python
plt.figure()

plt.subplots_adjust(wspace=0.6,hspace=0.6)
for i in np.arange(1,7):
    plt.subplot(2,3,i)

   
    

```

```python
plt.figure()

fig,ax=plt.subplots(2,3,sharex='col',sharey='row')
x=np.linspace(0,50)

ax[0,0].plot(np.sin(x))
plt.show()
```

```python
fig=plt.figure()

grid=plt.GridSpec(2,3,hspace=0.4,wspace=0.4)

fig.add_subplot(grid[0,0])
fig.add_subplot(grid[0,1:3])
```

```python
fig = plt.figure(figsize=(7, 9))
gs = plt.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])
ax=fig.add_subplot(gs[2:,:])
ax.set_aspect('equal')
```

```python
fig=plt.figure()

grid=plt.GridSpec(3,3,wspace=0.5,hspace=0.5)

main_plot=fig.add_subplot(grid[0:2,0:2])
right_plot=fig.add_subplot(grid[0:2,2])
down_plot=fig.add_subplot(grid[2,0:2],sharex=main_plot)

sns.boxplot(data.hwy,ax=right_plot,orient='v')
right_plot.set(ylabel='',yticks=[])
plt.setp(down_plot.get_xticklabels(),visible=False)
down_plot.tick_params(axis='x',length=0)
```
