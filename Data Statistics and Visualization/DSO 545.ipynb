```
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
%%time
pdc = pd.read_excel('Property Data Compiled.xlsx')
```

```
%%time
sdc = pd.read_excel('Sales Data Compiled.xlsx')
```

```
%%time 
keys = pd.read_excel('Key.xlsx')
```

```
df_keys = ["PID", "Owner Name", "Owner Address", "Number Of Stories",
        'Features', 'Avg Asking/SF', 'Avg Effective/SF', "SID","Market Time",
        "Listing Broker Company","Land Area AC", 'Age', 'Year Built', 'Building Materials',
        'PropID','SID','Net Income','Property State','Property City','Property Zip Code',
        'Secondary Type', 'Number Of 1 Bedrooms','Number Of 2 Bedrooms','Number Of 3 Bedrooms',
        'Number Of Other Bedrooms', 'Number Of Studios', 'Star Rating',
       'Owner Name', 'Owner Address', 'Floor Area Ratio', 'Number Of Stories', 'Number Of Floors', 'Latitude', 'Longitude', 'Building Status']
```

```
pdc_keys = []
for key in pdc.keys():
    if key in df_keys:
        pdc_keys.append(key)
```

```
sdc_keys = []
for key in sdc.keys():
    if key in df_keys:
        sdc_keys.append(key)
```

```
pdc = pdc[pdc_keys]
```

```
sdc = sdc[sdc_keys]
```

```
%%time
data = pdc.merge(keys, left_on = 'PropID', right_on= 'PID').merge(sdc, on = 'SID')
```

```
data.head()
```

```
data.shape
```

```
for key in data.keys():
    if '_y' in key:
        data.drop(columns=[key], inplace=True)
```

```
data.shape
```

Convert feature column to catogrical

```
%%time
data['Features'].fillna('None', inplace = True)
feature_list = data['Features'].apply(lambda x: np.array(list(map(lambda y: y[1:] if y[0] == ' ' else y, x.split(','))))).values
features = []
for row in feature_list:
    for feature in row:
        features.append(feature)

features = list(set(features))

for item in features:
    data[item] = data['Features'].apply(lambda x: 1 if item in x else 0)
```

Remove columns with a lot of 0s

```
%%time

feature_freq = pd.DataFrame({
    'Feature': features,
    'Percentage': list(map(lambda x: sum(data[x])/data.shape[0], features))
})
feature_freq.sort_values('Percentage', ascending = False, inplace = True)

for feature in features:
    if feature not in feature_freq[feature_freq['Percentage'] > 0.05]['Feature'].values:
        data.drop(columns=[feature], inplace = True)
```

```
prop = data[data['Building Status'] == 'Proposed']
data = data[(data['Building Status'] != 'Demolished') | (data['Building Status'] != 'Proposed')]
```

```
nan_zero = ['Number Of Studios_x', 'Number Of 3 Bedrooms_x', 'Number Of 1 Bedrooms_x', 'Number Of 2 Bedrooms_x', 'Number Of Other Bedrooms']
for key in nan_zero:
    data[key].fillna(0, inplace = True)
```

```
imp = data['Building Materials'].value_counts().index[0]
data['Building Materials'].fillna(imp, inplace = True)
```

```
data['Market Time'].min()
```

```
%%time
nan_freq = pd.DataFrame({
    'Column': data.columns,
    'Nan': list(map(lambda x: sum(data[x].isna())/data.shape[0], data.columns))
})
nan_freq.sort_values('Nan', ascending = False, inplace = True)
```

```
nan_freq
```

```
nan_keys = nan_freq.iloc[:3,:]['Column'].values
```

```
data.shape
```

```
for key in nan_keys:
    data.drop(columns=[key], inplace = True)
```
