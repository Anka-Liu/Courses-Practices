```python

html=urlopen('https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue')
bs1=BeautifulSoup(html,'html.parser')
```

```python
print(bs1.prettify())
```

```python
bs1.find("table",{"class":"wikitable"}).find_all("tr")[0].find_all('th')[0].text.replace('\n','')
```

```python
firstline=bs1.find("table",{"class":"wikitable"}).find_all("tr")[1].find_all('td')
```

```python
[col.text.replace('\n','').replace('\xa0','') for col in firstline]
```

```python
data=[]
for i in range(1,len(bs1.find("table",{"class":"wikitable"}).find_all("tr"))):
    line=bs1.find("table",{"class":"wikitable"}).find_all("tr")[i].find_all('td')
    data.append([col.text.replace('\n','').replace('\xa0','') for col in line])
```

```python
import pandas as pd

df=pd.DataFrame(data).iloc[:,1:]
```
