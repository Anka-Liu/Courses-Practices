```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('mpg(1)(1).csv')
data.head()
```

```python
datavis=data.groupby('manufacturer')['cty'].mean()

datavis=datavis.sort_values().reset_index()
datavis['manufacturer']=datavis['manufacturer'].str.title()
```

```python
plt.scatter(x=datavis['cty'],y=datavis['manufacturer'],color='firebrick',alpha=0.6);
ax=plt.gca()
ax.hlines(y=datavis['manufacturer'],xmin=10, xmax=26,color='grey',alpha=0.6,linestyle='-.',linewidth=0.8)
plt.title('Average car city mileage')
plt.xlabel('Mileage')

ax.tick_params(axis='both', which='both', length=0)
sns.despine(left=True,bottom=True)
plt.show()
```

```python
 
    
data['cty_z']=(data['cty']-data['cty'].mean())/data.cty.std()
```

```python
dataz=data.groupby('manufacturer')['cty_z'].mean().reset_index().sort_values('cty_z')
```

```python
dataz
```

```python
plt.figure()
ax1=plt.gca()

ax1.hlines(y=dataz.manufacturer,
          xmin=0,
          xmax=dataz.cty_z,
          color=['green' if x>0 else 'red' for x in dataz.cty_z],
          linewidth=3,
          alpha=0.6)

ax1.hlines(y=dataz.manufacturer,
          xmin=-2,
          xmax=2,
          color='grey',
          linewidth=0.8,
          alpha=0.4,linestyle=':')

ax1.tick_params(axis='both',length=0)
```
