```python
m=folium.Map(location=[37,-102],zoom_start=5)

folium.Choropleth(geo_data=states,
                  name='Chropleth',
                  data=unemployment,
                  columns=['State','Unemployment'],
                  key_on='feature.properties.id',
                  fill_color='Greens',
                  fill_opacity=0.7,
                  line_opacity=0.2)
```

```python
# * repeat 0 or more char
# + repeat 1 or more char
# {n} repeat n char
# . 1 char
# ? preceding optional
# ^ start match
# $ end match
# []  match the inside
# [^..] match any except the specified
# \d digit
# \w word
# \s space
#######

mytext="Dso DSO545 usc 5 Python usc"
pattern='usc'
re.findall(pattern,mytext)
```

```python
pattern='[a-z]'
re.findall(pattern,mytext)
```

```python
pattern='[A-Z]+'
re.findall(pattern,mytext)
```

```python
pattern='[A-Za-z]+'
re.findall(pattern,mytext)
```

```python
pattern='[0-9]+'
re.findall(pattern,mytext)
```

```python
mytext=["apple","banana","orange","avocado","cheeries","a","strawberries"]
pattern=r"^[aeiou].*"

for m in mytext:
    match=re.sub(pattern,"*****",m)
    print(match)
    
```

```python
mytext=["dso@gmail.com","dso.545","dso545@gmail.com","dso@gmail","545dso@gmail.com"]
pattern=r".+@.+\.com$"

for m in mytext:
    match=re.findall(pattern,m)
    print(match)
```
