1. (1 point) Use the panda’s dataframe method to to_datetime() to update the datat type of the Date
variable to a date/time variable.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trump= pd.read_csv('trump_approval.csv')
trump
```

```python
type(trump['Date'][0])
```

```python
trump['Date']=pd.to_datetime(trump['Date'])
trump['Date']
```

2. (1.5 points) Create a line chart that shows President’s Trump approval rating as follows. The approval
line is green, the disapproval line is orange, and the title is bold of fontsize 14. In addition, the plot’s
figure size is 11x6


```python
plt.figure(figsize=(11,6))
plt.plot('Date','Approve', data=trump, color='green',label='Approval')
plt.plot('Date','Disapprove',data=trump,color='orange',label='Disapproval')
plt.title('President Trump\'s Approval Rating',fontsize=14,weight='bold')
plt.legend(loc='upper left')
plt.ylabel('President\'s Rating(%)')
plt.show()
```

3. (1.5 points) Add a horizontal line at “2018-01-01”. The vertical line should be dashed and grey.


```python
plt.figure(figsize=(11,6))
plt.plot('Date','Approve', data=trump, color='green',label='Approval')
plt.plot('Date','Disapprove',data=trump,color='orange',label='Disapproval')
plt.axvline(x='2018-01-01',ls='--',c='grey')
plt.title('President Trump\'s Approval Rating',fontsize=14,weight='bold')
plt.legend(loc='upper left')
plt.ylabel('President\'s Rating(%)')
plt.show()
```

5. (1 points) Add a horizontal line at the 50% level approval as follows. The line should be dotted red.


```python
plt.figure(figsize=(11,6))
plt.plot('Date','Approve', data=trump, color='green',label='Approval')
plt.plot('Date','Disapprove',data=trump,color='orange',label='Disapproval')
plt.axvline(x='2018-01-01',ls='--',c='grey')
plt.axhline(y=50,ls=':',c='red')
plt.title('President Trump\'s Approval Rating',fontsize=14,weight='bold')
plt.legend(loc='upper left')
plt.ylabel('President\'s Rating(%)')
plt.ylim((30,70))
plt.show()
```

6. (1.5 points) Use the matplotlib method axvspan() to add a vertical span to show the government
shutdown period between Dec 22, 2018 and Jan 1, 2019. The span should be grey and the alpha level
should be 0.3.

```python
plt.figure(figsize=(11,6))
plt.plot('Date','Approve', data=trump, color='green',label='Approval')
plt.plot('Date','Disapprove',data=trump,color='orange',label='Disapproval')
plt.axvline(x='2018-01-01',ls='--',c='grey')
plt.axhline(y=50,ls=':',c='red')
plt.axvspan(xmin='2018-12-22',xmax='2019-01-25',facecolor='grey',alpha=0.3)
plt.title('President Trump\'s Approval Rating',fontsize=14,weight='bold')
plt.legend(loc='upper left')
plt.ylabel('President\'s Rating(%)')
plt.ylim((30,70))
plt.show()
```

```python
plt.figure(figsize=(11,6))
plt.plot('Date','Approve', data=trump, color='green',label='Approval')
plt.plot('Date','Disapprove',data=trump,color='orange',label='Disapproval')
plt.axvline(x='2018-01-01',ls='--',c='grey')
plt.axhline(y=50,ls=':',c='red')
plt.axvspan(xmin='2018-12-22',xmax='2019-01-25',facecolor='grey',alpha=0.3)
plt.text(x='2018-01-01',y=65,s='One year\nin office',fontsize=12)
plt.text(x='2018-12-01',y=65,s='Dec 22- Jan 25\nGovernment shutdown',fontsize=12)
plt.text(x='2019-08-01',y=57,s='57%\nDisapprove',fontsize=12,color='orange')
plt.text(x='2019-08-01',y=39,s='39%\nApprove',fontsize=12,color='green')
plt.title('President Trump\'s Approval Rating',fontsize=14,weight='bold')
plt.legend(loc='upper left')
plt.ylabel('President\'s Rating(%)')
plt.ylim((30,70))
plt.show()
```
