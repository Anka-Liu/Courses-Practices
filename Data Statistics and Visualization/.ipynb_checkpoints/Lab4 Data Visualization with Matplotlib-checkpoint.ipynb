```python
A=[1,2,3,4]
B=[1,3,4,2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(A,B)
```

```python
plt.figure(figsize=(8,4))
plt.plot(A,B)
```

```python
plt.figure(figsize=(8,4))
plt.plot(A,B,linestyle= '-.')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(A,B,linestyle= '--',marker='o',markersize=10,color='green',label='GreenLine')
plt.ylim(0,5)
plt.title('This is a title')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.xlabel('X Label', fontsize=12)
plt.ylabel('Y Label', fontsize=12)
plt.title('This is a title',fontsize=12)
plt.show()
```

```python
#4. 

C=[0,1,2,3]
D=[0.5,2,1,4]

plt.figure(figsize=(8,4))
plt.plot(A,B,linestyle= '--',marker='o',markersize=10,color='green',label='GreenLine')
plt.plot(C,D,linestyle='-.',marker='<',color='red',label='RedLine')
plt.ylim(0,5)
plt.title('This is a title')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
google=pd.read_csv('GOOG.csv')
```

```python
amazon=pd.read_csv('AMZN.csv')
```

```python
google.head()
```

```python
plt.plot(google.Date,google.Close,label='Google')
plt.plot('Date','Close',data=amazon,label='Amazon')

plt.title('GA stock trend 2018-2019')
plt.xticks(range(0,300,20),rotation=70)
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
iq_scores=np.random.normal(100,15,100)
```

```python
plt.hist(iq_scores)
plt.axvline(x = iq_scores.mean(),color='red')
plt.axvline(x=iq_scores.mean()+iq_scores.std(),color='red',linestyle='--')
plt.axvline(x=iq_scores.mean()-iq_scores.std(),color='red',linestyle='--')
```
