## Q5. Team Assignment

The following MIP is a simplified version of the program used to assign students into project teams in this course, to balance the overall characteristics of each team.

**Data:**

- $I$: set of students.
- $n$: number of teams
- $J=\{1,2,\cdots,n\}$ : set of teams.
- $K$: set of characteristics.
- $a_{ik}$: student $i$'s value for characteristics $k$.
- $w_k$: the weight for characteristics $k$ in the objective.
- $L_k$: the best lower bound for the sum of characteristic $k$ for any team. 
- $U_k$: the best upper bound for the sum of characteristics $k$ for any team.

You should assume that the data is given in a excel file with the same format as the `HWK8-Q5-input-1.xlsx` and `HWK8-Q5-input-2.xlsx` files attached to this assignment. 

The sheet named "Students" encodes $I$, $K$ and $a_{ik}$'s. In the below screenshot of `HWK8-Q5-input-1.xlsx`, $I=\{A,B,C,D,E,F\}$, and $K=\{Person, Male, Programmer, Math, Speaking\}$.

<img src='Hwk8-Q5-1a.png' width=500>

The sheet named "Parameters" encodes the $w_k$, $L_k$ and $U_k$ for each characteristic $k$.

<img src='Hwk8-Q5-1b.png' width=500>

**Decision variables:**

- $x_{ij}$ : whether to assign student $i$ to team $j$. (Binary)
- $s_k$,  and $t_k$ : auxilliary variables for each characteristic $k$. (Continuous)

**MIP:**

$$\begin{aligned}
\text{Minimize:} && \sum_{k \in K} w_k(s_k+t_k) \\
\text{subject to:} && \\
\text{(Every person assigned)} && \sum_{j \in J} x_{ij} & = 1 && \text{For each person $i \in I$.}\\
\text{(Team balance)} && L_k - s_k \le \sum_{i \in I} a_{ik}x_{ij} & \le U_k + t_k && \text{For each team $j \in J$ and each $k \in K$.} \\
\text{(Binary)} && x_{ij} & \in \{0, 1\} && \text{for all $i$ and $j$.}\\
\text{(Non-negativity)} && s_k, t_k & \ge 0 && \text{for all $k$.}
\end{aligned}$$

**Write a function called "assignTeams" with the following input arguments**:

- **inputFile:** path to the input spreadsheet.
- **n:** the number of teams to divide students into.

**The function should return two variables:**

- **df:** a DataFrame with one column called "Team". The index should be the name of each individual, and the column "Team" should specify the number $j$ to which the person is assigned.
- **objval:** the optimal objective value.

```python
def assignTeams(inputFile, n):
    data_name = pd.read_excel(inputFile, sheet_name='Students')
    data = data_name.iloc[:, 1:]
    params = pd.read_excel(inputFile, sheet_name='Parameters', index_col=0)
    mod = gb.Model()
    Bi = mod.addVars(len(data), n, vtype=gb.GRB.BINARY)
    Li = mod.addVars(len(params.columns), n)

    mod.setObjective(gb.quicksum(
        params.loc['Weights', :][i]*(Li[i, 0]+Li[i, 1]) for i in range(len(data.columns))))
    mod.addConstrs(Bi.sum(i, '*') == 1 for i in range(len(data)))
    for i in range(len(params.columns)):
        for j in range(n):
            mod.addConstr(params.loc['L'][i]-Li[i, 0] <= gb.quicksum(
                [data.iloc[k, i]*Bi[k, j] for k in range(len(data))]))
            mod.addConstr(gb.quicksum(
                [data.iloc[k, i]*Bi[k, j] for k in range(len(data))]) <= params.loc['U'][i]+Li[i, 1])
    mod.setParam('OutputFlag', False)
    mod.optimize()
    matrix = []
    for i in range(len(data)):
        for j in range(n):
            matrix.append(Bi[i, j].X)

    matrix = np.array(matrix).reshape(-1, n)
    output = pd.DataFrame(matrix, columns=list(range(1, n+1)))
    output['Names'] = data_name['Names']

    def FindTeam(x):
        for i in range(1, n+1):
            if x[i][0] == 1:
                return i
    df = pd.DataFrame(output.groupby(
        'Names').apply(FindTeam), columns=['Team'])
    objval = mod.objval
    return df, objval
```
