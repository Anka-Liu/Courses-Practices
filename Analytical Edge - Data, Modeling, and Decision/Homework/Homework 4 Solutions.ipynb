It is possible to render mathematical expressions nicely in Markdown cells, using a type of script called "LaTex". LaTex the standard format for people in most technical fields to write math nicely. (You can Google for examples of what you can do with LaTex.)

A single dollar sign is for mathematical expressions within the same line, and double dollar signs are for a standalone mathematical expressions in its own line. Hence, `$$X > 0.$$` is rendered as  $$X>0.$$

## Q2. DMD Exercise 2.4 

An oil company is drilling for oil at three promising sites. According
to geological tests, the probabilities of finding oil at these three sites are 0.70, 0.85, and 0.80,
respectively. The presence of oil at any one of the sites is presumed to be independent of the
presence of oil at any of the other sites.

(a) What is the probability of finding oil at all three of the sites?

(b) What is the probability of not finding oil at any of the three sites?

**Answer:**

**(a)** Let the event of finding oil at each of the three sites be $A$, $B$, and $C$ respectively. We have $P(A)=0.7$, $P(B)=0.85$ and $P(C)=0.8$. 

Since the events are independent, the probability of finding event at all three sites is

$$ P ( A \text{ and } B  \text{ and } C) = P(A)P(B)P(C) = (0.7)(0.85)(0.8) = 0.476. $$

**(b)** Define events $A'$, $B'$ and $C'$ as the not finding oil at each of the three sites respectively. We have $P(A')= 1-P(A) = 0.3$. Similarly, $P(B')=0.15$ and $P(C')=0.2$. Since the events $A$, $B$, $C$ are independent, we know that their complements $A'$, $B'$ and $C'$ must also be independent.

Therefore, the probability of not finding oil at any of the three sites is

$$ P ( A'  \text{ and } B'  \text{ and } C') = P(A')P(B')P(C') = (0.3)(0.15)(0.2) = 0.009. $$

Note: the above argument works for finding the probability any of the 8 combinations of the three events: $(A \text{ and } B \text{ and } C)$, $(A  \text{ and } B  \text{ and } C')$, $(A  \text{ and } B'  \text{ and } C)$, $(A  \text{ and } B'  \text{ and } C')$, $(A'  \text{ and } B  \text{ and } C)$, $(A'  \text{ and } B  \text{ and } C')$, $(A'  \text{ and } B'  \text{ and } C)$ and $(A'  \text{ and } B'  \text{ and } C')$.

## Q3. DMD Exercise 2.10 
A hardware store has received two shipments of halogen lamps. The first shipment contains 100 lamps, 4 percent of which are defective. The second shipment contains 50 lamps, 6 percent of which are defective. Suppose that Emanuel picks a lamp (at random) off of the shelf and purchases it, and he later discovers that the lamp he purchased is defective. Is the defective lamp more likely to come from the first shipment or from the second shipment?

**Answer:**

Intuitively speaking, the first shipment is expected to have $100 \times 0.04 = 4$ defective lamps, and the second shipment is expected to have $50 \times 0.06 = 3$ defectively lamps. Hence, a uniformly random defective lamp has a probability of $\frac{4}{7}$ of coming from the first shipment and a probability of $\frac{3}{7}$ of coming from the second shipment.

Formally, let $S1$ and $S2$ denote the event that a uniformly random lamp off the shelf comes from the first shipment and the second shipment respectively. Let $D$ denote the event that the lamp is effective. We know that 

$$P(S1) = \frac{100}{100+50} = \frac{2}{3},$$
$$P(S2) = \frac{50}{100+50} = \frac{1}{3},$$
$$P(D | S1) = 0.04,$$
$$P(D | S2) = 0.06.$$

Therefore,

$$P( S1  \text{ and } D) = P(S1)P(D | S1) = (\frac{2}{3})(0.04),$$
$$P( S2  \text{ and } D) = P(S2)P(D | S2) = (\frac{1}{3})(0.06),$$
$$P( D ) = P(S1  \text{ and } D) + P( S2  \text{ and } D) = (\frac{2}{3})(0.04)+(\frac{1}{3})(0.06).$$

By the definition of conditional probability,

$$P(S1 | D) = \frac{P(S1  \text{ and } D)}{P(D)} = \frac{\frac{2}{3}(0.04)}{\frac{2}{3}(0.04)+\frac{1}{3}(0.06)} = \frac{4}{7}. $$

Note that we could also directly obtain this using Bayes' rule:

$$P(S1 | D) = \frac{ P(D|S1)P(S1)}{P(D|S1)P(S1)+P(D|S2)P(S2)} =  \frac{(0.04)(\frac{2}{3})}{(0.04)(\frac{2}{3})+(0.06)(\frac{1}{3})}  =\frac{4}{7}.$$

## Q4. DMD Exercise 2.11 
It is estimated that one third of the population in a given county is infected with the tuberculosis (TB) bacteria. The human body is usually able to successfully fight the TB bacteria and so prevent the onset of the TB disease. Consequently, a person infected with the TB bacteria has only a 10 percent chance of development the TB disease over his/her lifetime.

(a)	Suppose that we choose a person at random from all of the people in the county. What is the probability that this person will have TB disease over his/her lifetime?

(b)	Suppose that among those people who have died in the county in the last year, we perform an autopsy on one of the bodies chosen at random, and we find that this person did not have the TB disease. What is the probability that this person had been infected with the TB bacteria?

**Answer:**

**(a)** Since a third of people have the TB bacteria and 10\% of such people develop the disease in their lifetime, the probability that a randomly chosen person develops the diease in his/her life time is $\frac{1}{10} \times \frac{1}{3} = \frac{1}{30}$.

**(b)** There are two reasons a person does not have the TB disease: 1) the person was not infected with the TB bacteria; 2) the person was infected but fought it off. The probability of the first event is $\frac{2}{3}$ and the probability of the second event is $\frac{1}{3} \times \frac{9}{10} = \frac{3}{10}$. Therefore, conditional on the person not having the disease, the probability that the person has been infected with the TB bacteria is 

$$ \frac{ 3/10 } { 2/3 + 3/10 } = \frac{9}{29}.$$

To express the above argument using math notation, define $B$ to be the event that a person is infected with the TB bacteria, and $D'$ to be the event that a person does not have the disease. Let $B'$ be the complement of $B$, meaning that $B'$ denotes the event that the person is not infected with the bacteria. We have, 

$$P(B) = 1/3,$$
$$P(B') = 1-P(B) = 2/3,$$
$$P(D'|B) = 9/10.$$
$$P(D'|B') = 1.$$

Therefore, by Bayes' rule,
$$P(B|D') = \frac{P(D'|B)P(B)}{P(D'|B)P(B)+P(D'|B')P(B')} = \frac{(9/10)(1/3)}{(9/10)(1/3)+(1)(2/3)} = \frac{9}{29}.$$

## Q5. (Geographic Testing)

Extensive logging has exposed a hillside in San Carlos to the possibility of a mudslide. The mayor has consulted a geographic expert, who estimates that there is only a 10\% chance that the slide will occur within the next year. The expert adds that roughly 5\% of such slides would break through the existing retaining wall, which would cause a damage of one million dollars. (The slides that don't break the wall would cause negligible damage.) 

The expert also mentioned that she can conduct a geological test to better assess the likelihood of a mud slide. The test has three possible outcomes, as seen in the table below, which also summarize past data (from other locations) on whether a subsequent slide indeed happened afterward.

| Test Outcome | Slide within 1 year | No slide within 1 year | 
|--|--|--|
|Positive| 900 | 50 | 
|Neutral| 25 |  100 |
|Negative| 75 | 850 |

**(a)** Suppose that the expert's initial estimate of 10\% chance is correct, what's the probability that the test would yield a positive result in San Carlos?


The following solutions illustrates one way of approaching a more challenging math problem, which is 1) to describe what is desired and given; 2) to identify a plan of going from the given to the desired; and 3) execute the plan and perform the actual computations. 

**Describe:**

Desired: $P(positive)$.

Given: $P(slide)$, and past data on test outcomes and slides. 

**Identify/Plan:**

Note that we cannot directly use the above table and do $950/(950+125+925)$, because the prior probability of a mud slide in San Carlo may be different from those in the past data.

However, using the past data, we can compute $P(positive|slide)$, $P(neutral|slide)$, $P(negative|slide)$, as well as $P(positive|slide')$ ($slide'$ denotes no slide), $P(neutral|slide')$ and $P(negative|slide')$.

We can use the joint probability table or conditional probability to compute $P(slide \text{ and } positive)$, and $P(slide' \text{ and } positive)$, as well as their sum, $P(positive)$.

**Execute:** 

Using the past data, we have
$$\begin{aligned}
P(positive | slide) &= 0.9,\\
P(neutral | slide) &=0.025,\\
P(negative | slide) &= 0.075, \\
P(positive | slide')&=0.05,\\
P(neutral | slide')&=0.1,\\
P(negative | slide')&=0.85.
\end{aligned}$$

Therefore,

$$P(positive \text{ and } slide) = P(slide)P(positive|slide)=(0.1)(0.9)=0.09,$$
$$P(positive \text{ and } slide') = P(slide')P(positive|slide')=(0.9)(0.05)=0.045,$$
$$P(positive)=0.09+0.045=0.135.$$

**(b)** If there is a way to strengthen the retaining wall so that it would not break under any mud slide, how much should the mayor be willing to pay for the wall supposing that the geographic test yields a positive result? What about if it yields a neutral result?

**Describe:**

Desired: the expected benefit of strengthening the wall conditional on the test result being positive; analogous expected benefit conditional on the test result being neutral.

Given: the joint probabilities as in part a), the probability that a slide breaks the wall without strengthening, and the damage of a broken wall.

**Identify/Plan:**

The mayor should be able to pay amount that is equal the expected damage that can be avoided.

$$E[damage | slide ] =(0.05)(1000000)=50000.$$

Without strengthening the wall,

$$E[damage | positive] = E[damage | slide \text{ and } positive]P(slide | positive)+E[damage | slide' \text{ and } positive]P(slide' | positive),$$

which simplifies to

$$E[damange| positive ] = E[damage| slide]P(slide|positive) + 0.$$

We can use the joint probabilities from part a) to compute $P(slide|positive)$.

**Execute:**

Using the joint probabilities from part $a)$,

$$P(slide | positive) = \frac{P(slide \text{ and }positive)}{P(slide \text{ and }positive)+P(slide' \text{ and }positive)}= \frac{0.09}{0.09+0.045} = \frac{2}{3}.$$

$$P(slide | neutral) = \frac{P(slide \text{ and }neutral)}{P(slide \text{ and }neutral)+P(slide' \text{ and }neutral)}= \frac{(0.1)(0.025)}{(0.1)(0.025)+(0.9)(0.1)} = \frac{1}{36}.$$


Without strengthening the wall, the expected damage (conditional on a positive test result) is 

$$E[damage| positive ] = (0.05)(1000000)P(slide|positive) \approx 33333.$$
$$E[damage| neutral ] = (0.05)(1000000)P(slide|neutral) \approx 1389.$$

Therefore, the mayor should be willing to pay about 33K for strengthening the wall given a positive test result, and only 1.4K given a neutral result.

**(c)** Suppose the way to strengthen the retaining wall costs $20000$ dollars, how much should the mayor be willing to pay for the geological test?

**Describe:**

Desired: willingness to pay for the test.

Given: the cost of building a retaining wall, the expected damage without the wall conditional on the test result.

**Identify:**

The value of the test comes from being able to make better decision from it.

Knowing the test result is related to conditional probability.

Given whatever information, the decision we need to make is to choose between doing nothing and strengthening the wall.

**Plan:**

Compare what can be done without the test to what can be done with the test. 

What can be done without the test is the maximum of zero (doing nothing) and strengthening the wall (which yields the damage saved minus the cost of the wall).

What can be done with the test is to be able to condition on the test result when making this decision. For example, from the above, the mayor is only willing to pay more than $20000$ for the test if there is a positive result. (The expected damage under a neutral and negative result are both less than $20000$). Hence, when the test is positive, the mayor saves $33333-20000=13333$ dollars. When the test is neutral or negative, the mayor would not build the retaining wall, and there are no savings from knowing the test result. 

**Execute:**
Without doing the test, the mayor would not strengthen the wall, because $E[damage]=50000P(slide)=5000< 20000$.

The value of the test comes from being able to better make the decision about strengthening the wall. Precisely speaking, the total value from the test is

$$\max(E[damage|positive]-20000,0)P(positive)+\max(E[damage|neutral]-20000,0)P(neutral)+\max(E[damage|negative]-20000,0)P(negative),$$

which simplifies to

$$(13333)(0.135)+0+0 \approx 1800.$$

Hence, the mayor should be willing to pay 1800 for the test.
