```python
import pandas as pd
import scipy.stats as sps
import numpy as np
```

```python
mydata=pd.read_csv("training_testing.csv")
```

```python
mydata.head()
```

```python
stack=[0]*len(mydata.columns)
KSFDR=pd.DataFrame({"KS":stack,"FDR":stack,"KS rank":stack,"FDR rank":stack,"Average rank":stack})
```

```python
goods=mydata[mydata['fraud_label']==0]
bads=mydata[mydata['fraud_label']==1]
```

```python
goods.head()
```

```python
i=0 
for column in mydata:
    ks=sps.ks_2samp(goods[column],bads[column])
    KSFDR.loc[i,'KS']=ks[0]
    print(ks)
    i=i+1
```

```python
topRows=int(round(len(mydata)*0.03))
j=0
for column in mydata:
    temp1=mydata.sort_values(column,ascending=False).head(topRows)
    temp2=mydata.sort_values(column,ascending=True).head(topRows)
    needed1=temp1.loc[:,'fraud_label']
    needed2=temp2.loc[:,'fraud_label']
    FDR1=sum(needed1)/len(bads)
    FDR2=sum(needed2)/len(bads)
    FDRate=np.maximum(FDR1,FDR2)
    KSFDR.loc[j,'FDR']=FDRate
    j+=1
```

```python
KSFDR['KS rank']=KSFDR['KS'].rank(ascending=True)
KSFDR['FDR rank']=KSFDR['FDR'].rank(ascending=True)
KSFDR['Average rank']=(KSFDR['KS rank']+KSFDR['FDR rank'])/2
```

```python
KSFDR.sort_values(by=['KS rank'],ascending=False).to_csv("KSFDR_KS.csv")
```

```python
KSFDR.sort_values(by=['FDR rank'],ascending=False).to_csv("KSFDR_FDR.csv")
```
